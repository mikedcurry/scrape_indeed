```
import requests
import bs4
from bs4 import BeautifulSoup as bs

import pandas as pd
import numpy as np
import time
```

```
# Importing the sample page again:

url = 'https://www.indeed.com/jobs?q=data+scientist&l=Pittsburgh%2C+PA'
page = requests.get(url)

broth = bs(page.text, 'html.parser')
```

```
# just looking at it.

print(broth.prettify())
# 'it was too long' - someone said.
```

```
# Function to pull out job titles from the above static page...

def get_title(broth):
    jobs = []
    for div in broth.find_all(name='div', attrs={'class': 'row'}):
        for a in div.find_all(name='a', attrs={'data-tn-element': 'jobTitle'}):
            jobs.append(a['title'])
    return(jobs)
```

```
get_title(broth)
```

```
# Func to get company names:

def get_company(broth):
    companies = []
    for div in broth.find_all(name='div', attrs={'class': 'row'}):
        for company in div.find_all(name='span', attrs={'class': 'company'}):
            if len(company) > 0:
              for nut in company:
                companies.append(nut.text.strip())
            else:
              sec_try = div.find_all(name='span', attrs={'class':'result-link-source'})
                for span in sec_try:
                  companies.append(span.text.strip())
    return(companies)


```
